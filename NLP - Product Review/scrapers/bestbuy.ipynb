```python
'''
This is to retreive the reviews from BestBuy
Galaxy s9 and iPhone X

written by Soi Kang
3/27/2019
'''
import os
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

#read the html files in the folder
PATH = 'bestbuy'

# list of html files to read
htmls = []

#get the list of the files
for html in os.listdir(PATH):
    htmls.append(html)

dfs = []
i=0
for html in htmls:
    
    #read it as a Soup object
    try :
        soup = BeautifulSoup(open(os.path.join(PATH,html)),"html.parser")
    
    except UnicodeDecodeError :
        soup = BeautifulSoup(open(os.path.join(PATH,html), encoding='utf-8'),"html.parser")
    
    reviews = []
    for element in soup.find_all(class_= "ugc-review-body body-copy-lg"):

        review = element.find("p").get_text().strip()
        reviews.append(review)
    
    ratings = []
    for element in soup.find_all(class_= "c-review-average"):

        rating = element.get_text().strip()
        ratings.append(rating)
    
    labels = ['review','rating']
    df = pd.DataFrame({"review" : reviews, "rating" : ratings})
    df['source'] = "BestBuy"
    
    if "galaxy" in html :
        df['phone'] = "galaxy"
    else :
        df['phone'] = "iphone"
    
    dfs.append(df)
    print(i)
    i+=1
    
    
final = pd.concat(dfs)   
    
#reorder columns
cols = ['source','phone','review','rating']
final = final.reindex(columns=cols)        

```
