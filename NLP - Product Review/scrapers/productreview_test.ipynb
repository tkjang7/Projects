```python
import urllib.request
from bs4 import BeautifulSoup

#header information wjere you specify 
#pass them as header
#or selenium

# this is to scrape reviews from bestbuy for galaxy s9
# first review page
url = "https://www.productreview.com.au/listings/samsung-galaxy-s9?page=" +"#reviews"
headers = {'user-agent': 'my-app/0.0.1'}

request = urllib.request.Request(url, headers=headers)
response = urllib.request.urlopen(request)
html = response.read()
html = html.decode('utf-8')
    
# save the file
with open('bestbuyreviews_galaxy1.html', 'w') as new_file:
    new_file.write(html)
    
    
i=2
pages=86
while(i<pages):
    url="https://www.bestbuy.com/site/reviews/samsung-galaxy-s9-64gb-sunrise-gold-unlocked/6253419?page={}"
    url=url.format(i)
    request = urllib.request.Request(url, headers=headers)
    response = urllib.request.urlopen(request)
    html = response.read()
    html = html.decode('utf-8')
    
    # save the file
    with open('bestbuyreviews_galaxy'+str(i)+'.html', 'w') as new_file:
        new_file.write(html)
    
    i=i+1
    print(i)
```
